```python
from ipymuvue.widgets import VueWidget
from traitlets import Unicode, Dict, Int, List, Float, Tuple


class Widget(VueWidget):
    def __init__(self, edges):
        VueWidget.__init__(self, template=self.template, components=self.components)
        self.edges = edges

    template = r"""
        <planar-map :edges="edges" :width="width" :height="height" />
    """

    width = Int(1024R).tag(sync=True)
    height = Int(1024R).tag(sync=True)

    # for each edge, a sequence of control points: [[(x, y), ...], ...]
    edges = List(List(Tuple(Float(), Float()))).tag(sync=True)
    
    components = {
        "planar-map": open("demo/PlanarMap.vue"),
    }

def randfloat(a, b):
    import random
    return float(random.random() * (b - a) + a)

def edges(n, length=[4, 8], start=[0, 3], step=[-1, 1]):
    import random
    for i in range(n):
        edge = [(randfloat(*start), randfloat(*start))]
        for l in range(random.randint(*length) - 1):
            edge.append((edge[-1][0] + randfloat(*step), edge[-1][1] + randfloat(*step)))
        yield edge

widget = Widget(list(edges(16)))
widget
```
